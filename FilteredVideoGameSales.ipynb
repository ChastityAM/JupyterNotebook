```python
import csv
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
save_folder =(r"C:\Users\RAC\Documents\charts")
save_folder
```

```python
df = pd.read_csv(r"C:\Users\RAC\Documents\videogames-filtered.csv")

df
```

```python
sales_df = df[["Name","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
sales_df
```

```python
games_by_company_df = df.groupby(['Platform','Year'],as_index=False)['Name'].count()
games_by_company_df
```

```python
games_on_platform_df = df.groupby(['Publisher'],as_index=False)['Name'].count()
games_on_platform_df
```

```python
genre_perYear_df = df.groupby(['Genre','Year'],as_index=False)['Name'].count()
genre_perYear_df
```

```python
platform_yearlysales_df = df.groupby(['Platform','Year'],as_index=False)['Global_Sales'].sum()
platform_yearlysales_df
```

```python
platform_sales_df = df.groupby(['Platform'],as_index=False)['Global_Sales'].sum()
platform_sales_df
```

```python
platform_sales_df.plot.bar(x='Platform', y='Global_Sales')
plt.legend(loc="best")
plt.title("Amount of Games Sold On Each Gaming Console")
plt.xlabel("Console")
plt.ylabel("Sales")
plt.grid()

plt.savefig(save_folder + "/platform_sales_chart.png")
plt.show()
```

```python
genre_sales_df = df.groupby(['Genre'],as_index=False)['Global_Sales'].sum()
genre_sales_df
```

```python
genre_sales_df.plot(x='Genre', y='Global_Sales',figsize=(20,20))
plt.xticks(ticks=np.arange(12),labels=genre_sales_df['Genre'])
plt.legend()
plt.title("Amount of Games Sold In Each Genre")
plt.xlabel("Genres")
plt.ylabel("Sales")
plt.grid()

plt.savefig(save_folder + "/genre_sales_chart.png")
plt.show()
```

```python
year_sales_df = df.groupby(['Year'],as_index=False)['Global_Sales'].sum()
year_sales_df
```

```python
year_sales_df.plot.scatter(x='Year', y='Global_Sales')
plt.title("Global Videogame Sales from 1984 to 2016")
plt.savefig(save_folder + "/year_sales_chart.png")
plt.show()
```
