```python
import pandas as pd
import numpy as np
```

```python
df_original = df.shape[0]
df_original
```

```python
complete_df = df.dropna(subset= ["amount", "items"])
complete_df = complete_df[(complete_df["amount"] > 0) & (complete_df["items"] > 0)]
complete_df
```

```python
altered_df = complete_df.shape[0]
altered_df
```

```python
df_difference = len(df) - len(complete_df)
df_difference
```

```python
amount_average = complete_df["amount"].mean()
amount_average
```

```python
amount_total = complete_df["amount"].sum()
amount_total
```

```python
item_average = complete_df["items"].mean()
item_average
```

```python
item_total = complete_df["items"].sum()
item_total
```

```python
total_cust_df = complete_df[['customer','amount', 'items']].groupby('customer').sum()
total_cust_df = total_cust_df.rename(columns = {"amount": "Total Amount", "items": "Total Items"})
total_cust_df
```

```python
avg_cust_df = complete_df[['customer','amount', 'items']].groupby('customer').mean()
avg_cust_df = avg_cust_df.rename(columns = {"amount": "Average Amount", "items": "Average Items"})
avg_cust_df
```

```python
cust_df = pd.merge(total_cust_df,avg_cust_df, on=["customer"])
cust_df = cust_df[["Total Amount", "Average Amount","Total Items", "Average Items"]]
cust_df
```

```python
sorted_df = cust_df.sort_values("Total Amount", ascending = False)
sorted_df
```

```python
sorted_df = cust_df.sort_values("Average Amount", ascending = False)
sorted_df
```

```python
sorted_df["Average Amount"] = sorted_df["Average Amount"].round(2)
sorted_df
```

```python
sorted_df["Average Amount"] = sorted_df["Average Amount"].map(lambda x: f"${x}")
sorted_df
```

```python
names = ["Bronze", "Silver", "Gold", "Platinum"]
bins = [400, 800, 1200, 1600, 2000]

binned_df = pd.cut(sorted_df["Total Amount"], bins, labels = names)
binned_df
```
